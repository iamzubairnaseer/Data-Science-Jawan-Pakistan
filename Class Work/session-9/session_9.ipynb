### df.BasePay = df.BasePay.replace('Not Provided', 0, inplace=True).astype('float') --> ERROR!

Kyun Error Aaya?
1. .replace() mein inplace=True hota hai, lekin wo wahi par kaam khatam kar deta hai aur "None" return karta hai.
2. Jab aap aage .astype() lagate hain, to wo "None" par chalne ki koshish karta hai, jis se error aata hai.
3. .astype() ka apna koi inplace parameter nahi hota.
