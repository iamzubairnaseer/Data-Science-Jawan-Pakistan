### Q No.3 Display current date and time

```python
import datetime
```

```python
current_date = datetime.datetime.now().strftime("%D")
current_date
```

```python
current_date_2 = datetime.datetime.now().strftime("%d/%b/%Y")
current_date_2
```

```python
current_date_3 = datetime.datetime.now().strftime("%d/%m/%Y")
current_date_3
```

```python
now = datetime.datetime.now()
```

```python
current_time = now.strftime("%I:%S %p")
current_time
```

```python
current_time_2 = now.strftime("%H:%M:%S")
current_time_2
```
